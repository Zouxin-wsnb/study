```python
import pandas as pd
import numpy as np

df=pd.read_csv('simplified_house_prices.csv')
df.head()
df.info()
df.isnull().sum()
```

```python
df.fillna(df.median(), inplace=True)
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
print(df)
```

```python
import matplotlib.pyplot as plt

df[['SalePrice', 'LivingArea', 'GarageSize']].hist(bins=20,figsize=(15,8))

plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))
df.boxplot('SalePrice',ax=ax[0])
df.boxplot('LivingArea',ax=ax[1])
ax[0].set_title('SalePrice')
ax[1].set_title('LivingArea')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['LivingArea'],df['SalePrice'],alpha=0.6)
plt.xlabel('LivingArea')
plt.ylabel('SalePrice')
plt.title('SalePrice vs LivingArea')
plt.show()

plt.figure(figsize=(12,6))
plt.scatter(df['GarageSize'],df['SalePrice'],alpha=0.6)
plt.xlabel('GarageSize')
plt.ylabel('SalePrice')
plt.title('SalePrice vs GarageSize')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.regplot(x='LivingArea',y='SalePrice',data=df, scatter_kws={'alpha':0.6})
plt.xlabel('LivingArea')
plt.ylabel('SalePrice')
plt.title('SalePrice vs LivingArea add trend line')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.boxplot('SalePrice',by='Quality',grid=False)
plt.suptitle('')
plt.title('SalePrice Distribution by Quality')
plt.xlabel('Quality')
plt.ylabel('SalePrice')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0,0].hist(df['SalePrice'], bins=20)
ax[0,0].set_title('SalePrice_Histogram')

ax[0,1].hist(df['LivingArea'], bins=20)
ax[0,1].set_title('LivingArea_Histogram')

ax[1,0].scatter(df['LivingArea'], df['SalePrice'], alpha=0.6)
ax[1,0].set_title('SalePrice vs LivingArea')    

ax[1,1].scatter(df['GarageSize'], df['SalePrice'], alpha=0.6)
ax[1,1].set_title('SalePrice vs GarageSize')

plt.tight_layout()
plt.show()
```

```python
df_select = df[['SalePrice', 'LivingArea', 'GarageSize']]
sns.pairplot(df_select)
plt.show()
```

```python
corr=df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```
