# Data Visualization using Matplotlib and Seaborn

This notebook covers various data visualization techniques using Python's Matplotlib and Seaborn libraries. We'll use a sample dataset from an electronics store to illustrate monthly sales, customer ratings, and product category breakdowns. Each section contains examples and exercises for practice.

## 0. Data Preparation

```python
import pandas as pd
import numpy as np

# Create sample data
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = np.random.randint(200, 500, size=12)
customer_ratings = np.random.uniform(3.5, 5.0, size=12).round(2)
product_categories = np.random.choice(['Laptops', 'Mobiles', 'Tablets'], size=12)

# Create DataFrame
df_sales = pd.DataFrame({
    'Month': months,
    'Sales': sales,
    'Customer_Rating': customer_ratings,
    'Product_Category': product_categories
})

# Display the first few rows
df_sales.head()
```

## 1. Basic Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Example 1: Line Plot for Monthly Sales**

```python

plt.figure(figsize=(10, 5))
plt.plot(df_sales['Month'], df_sales['Sales'],'o-b' )#marker='o', linestyle='-', color='b'
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

**Exercise 1**: Create a line plot showing the trend of `Customer_Rating` over the months.

```python
plt.figure(figsize=(10, 5))
plt.plot(df_sales['Month'], df_sales['Customer_Rating'], marker='s', linestyle='-', color='r')
plt.title('Monthly Customer Rating Trend')
plt.xlabel('Month')
plt.ylabel('Customer Rating')
# plt.grid(True)
plt.show()
```

**Example 2: Scatter Plot for Sales vs. Customer Ratings**

```python

plt.figure(figsize=(8, 5))
plt.scatter(df_sales['Sales'], df_sales['Customer_Rating'], color='g')
plt.title('Sales vs Customer Ratings')
plt.xlabel('Sales')
plt.ylabel('Customer Rating')
plt.show()
```

**Exercise 2**: Customize the scatter plot by adding grid lines and changing the color of the points to red.

```python

plt.figure(figsize=(8, 5))
plt.scatter(df_sales['Sales'], df_sales['Customer_Rating'], color='r')
plt.title('Sales vs. Customer Ratings')
plt.xlabel('Sales')
plt.ylabel('Customer Rating')

plt.grid(True)
plt.show()
```

## 2. Customizing Matplotlib Plots

**Example 3: Customizing a Bar Chart**

```python

plt.figure(figsize=(12, 6))
plt.bar(df_sales['Month'], df_sales['Sales'], color='skyblue')
plt.title('Monthly Sales with Customization')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=30)
plt.annotate('Highest Sales', xy=('Jul', df_sales.loc[df_sales['Month'] == 'Jul', 'Sales'].values[0]),
             xytext=('May', 450), arrowprops=dict(arrowstyle='->', color='red'))
plt.show()
```

**Exercise 3**: Create a customized line plot of `Customer_Rating` with annotations for the highest and lowest ratings.

```python
plt.figure(figsize=(12, 6))
plt.bar(df_sales['Month'], df_sales['Customer_Rating'], color='skyblue')
plt.title('Monthly Customer_Rating with Customization')
plt.xlabel('Month')
plt.ylabel('Customer_Rating')
plt.xticks(rotation=45)
plt.annotate('Highest Customer_Rating', xy=('Jun', df_sales.loc[df_sales['Month'] == 'Jul', 'Customer_Rating'].values[0]),
             xytext=('Apr', 6), arrowprops=dict(arrowstyle='->', color='red'))
plt.annotate('Lowest Customer_Rating', xy=('May', df_sales.loc[df_sales['Month'] == 'May', 'Customer_Rating'].values[0]),
             xytext=('Apr', 5), arrowprops=dict(arrowstyle='->', color='yellow'))
plt.show()
```

## 3. Seaborn Basics

```python
import seaborn as sns
```

**Example 4: Histogram and KDE Plot**

```python

plt.figure(figsize=(10, 5))
sns.histplot(df_sales['Sales'], kde=True, color='purple')
plt.title('Distribution of Monthly Sales')
plt.xlabel('Sales')
plt.show()
```

**Exercise 4**: Plot a histogram of `Customer_Rating` with a KDE overlay.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_sales['Customer_Rating'], kde=True, color='b')
plt.title('Distribution of Monthly Customer_Rating')
plt.xlabel('Customer_Rating')
plt.show()
```

## 4. Advanced Seaborn Visualizations

**Example 5: Categorical Plot (`sns.catplot`)**

```python

sns.catplot(x='Month', y='Sales', hue='Product_Category', data=df_sales, kind='bar', height=6, aspect=2)
plt.title('Monthly Sales by Product Category')
plt.show()
```

**Exercise 5**: Create a `catplot` to show the average `Customer_Rating` by month, using the `hue` parameter for `Product_Category`.

```python

sns.catplot(x='Month', y='Customer_Rating', hue='Product_Category', data=df_sales, kind='bar', height=6, aspect=2)
plt.title('Monthly Customer_Rating by Product Category')
plt.show()
```

## 5. Combining Matplotlib and Seaborn

**Example 6: Overlaying Seaborn and Matplotlib**

```python

plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Sales', data=df_sales, alpha=0.6, color='lightblue')
plt.plot(df_sales['Month'], df_sales['Customer_Rating'] * 100, marker='o', color='orange')  # Adjust scale for comparison
plt.title('Overlay of Sales and Customer Ratings')
plt.xlabel('Month')
plt.ylabel('Sales / (Customer Rating * 100)')
plt.legend(['Customer Rating (scaled)', 'Sales'])
plt.show()
```
